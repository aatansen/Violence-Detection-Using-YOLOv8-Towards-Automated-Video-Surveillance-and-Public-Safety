```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the path to the main directory
path = "/content/drive/MyDrive/thesis/dataset/roboflow"

# Define the subdirectories
subdirs = ["train", "test", "valid"]

# Function to get a random sample of images from a directory
def show_sample_images(directory, num_samples=2):
    image_files = os.listdir(directory)

    # Filter images starting with "V" or "NV"
    violence_files = [file for file in image_files if file.startswith("V")]
    non_violence_files = [file for file in image_files if file.startswith("NV")]

    # Get a random sample of images
    sample_files = random.sample(violence_files, num_samples // 2) + random.sample(non_violence_files, num_samples // 2)

    plt.figure(figsize=(12, 6))
    plt.suptitle(f"Sample Images from {directory}")

    for i, file in enumerate(sample_files, 1):
        img_path = os.path.join(directory, file)
        img = mpimg.imread(img_path)

        # Determine the label based on the filename
        label = "Violence" if file.startswith("V") else "Non-Violence"

        plt.subplot(1, num_samples, i)
        plt.imshow(img)
        plt.title(label)
        plt.axis("off")

# Loop through subdirectories and show sample images
for subdir in subdirs:
    images_dir = os.path.join(path, subdir, "images")
    show_sample_images(images_dir)

plt.show()
```

```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

# Set the path
path = "/content/drive/MyDrive/thesis/dataset/roboflow"

# Define the labels
labels = ['Non-violence', 'Violence']

# Function to get a random sample from a directory
def get_random_sample(directory, label, sample_size=2):
    images_path = os.path.join(path, directory, 'images')
    image_files = [f for f in os.listdir(images_path) if f.endswith('.jpg') and f.startswith(label[0])]
    random_sample = random.sample(image_files, min(sample_size, len(image_files)))
    return [(os.path.join(images_path, img), label) for img in random_sample]

# Get random samples from the 'train' directory
samples = []

# 'train', 'test', 'valid'
dir_to_show='train'
for label in labels:
    samples.extend(get_random_sample(dir_to_show, label))

# Display the images horizontally using matplotlib
fig, axes = plt.subplots(1, len(samples), figsize=(15, 5))

for i, (img_path, label) in enumerate(samples):
    img = Image.open(img_path)
    axes[i].imshow(img)
    axes[i].set_title(f"{label}")
    axes[i].axis('on')

# Add a subtitle to the entire figure at the bottom
fig.suptitle(f"Sample images from {dir_to_show} dataset", fontsize=13, y=0.17)

plt.show()
```
